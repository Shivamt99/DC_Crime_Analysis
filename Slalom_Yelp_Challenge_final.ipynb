*User Prespective (Yelp user data)*

Let's subset attributes related to user in a new dataframe.

- Since user name can be associated with multiple id's lets join id with user name to have a unique column to do futher analysis.

- Also, the column user_years_elite has the years a user was a part of elite squad lets map it to users who are elite or not

- Add all compliments column to a single column

```python
pd.options.mode.chained_assignment = None 

df_yelp_user = df_yelp_clean[['user_id','user_name','user_review_count','user_average_stars','user_years_elite','user_fans','user_yelping_since','user_votes_cool','user_votes_funny','user_votes_useful','user_compliments_cool', 'user_compliments_cute', 'user_compliments_funny', 'user_compliments_hot', 'user_compliments_list', 'user_compliments_more', 'user_compliments_note','user_compliments_photos','user_compliments_plain','user_compliments_profile','user_compliments_writer']]

# Creating new column user_id_name
df_yelp_user["user_name_id"] = df_yelp_user["user_name"].astype(str) + ' (' + df_yelp_user["user_id"].astype(str)+ ')'
```

### Most Impactful users 

Potential candidates to reach out to by HR team or to form a meetup/startup group with

```python
# function:  Enter the number of records it's should query to do analysis on the common user's found among the three atrributes
from IPython.display import display
def user_best(num):
    df_user_rc = df_yelp_user[['user_name_id','user_review_count','user_fans', 'user_average_stars', 'user_votes_useful','elite_user', 'user_all_compliments']].nlargest(num,'user_review_count',keep='all')
    df_user_fan = df_yelp_user[['user_name_id','user_review_count','user_fans', 'user_average_stars', 'user_votes_useful', 'elite_user','user_all_compliments']].nlargest(num,'user_fans',keep='all')
    df_user_votes = df_yelp_user[['user_name_id','user_review_count','user_fans', 'user_average_stars', 'user_votes_useful', 'elite_user','user_all_compliments']].nlargest(num,'user_votes_useful',keep='all')
    df_cor = df_user_rc.merge(df_user_fan,on='user_name_id').merge(df_user_votes,on='user_name_id')
    df_cor = df_cor[['user_name_id','user_review_count','user_fans', 'user_average_stars', 'user_votes_useful', 'elite_user','user_all_compliments']]
    display(df_cor)
    print("\n Total counts: "+ str(len(df_cor)))
```

```python
# Lets checkout the most impactful users

user_best(50)
```

After researching about the Yelp elite member conditions - it all makes sense now. ### Correlations

- Does having more review counts correlates with having high number of fans?
- Does having more useful votes given means the user get more compliments?

```python
# Group by business city, categories and get counts of each category skipping restaurant category

df_temp = df_biz_all[df_biz_all.business_categories!='restaurants'].\
    groupby(['business_city','business_categories']).\
    agg(business_categories_count=('business_categories', 'count'))
df_temp = df_temp.reset_index()
df_temp
```

```python
# Creating another column to get rank of of the category counts, using "first" it would assign sequential rank
# irrespective of dupicates. ( Note: this can also be approached using value_counts() which would automatically
# rank based on desc)

df_temp = df_temp.reset_index()
df_temp['category_rank'] = df_temp.groupby(["business_city"])["business_categories_count"].\
                 rank("first", ascending=False)
df_temp
```

```python
# Defining a fuction to filter top N categories.
#mFor demonstration purpose lets pick just top 3 categories from each city

def cat_rank(n):
    df_temp_topn = df_temp.query('category_rank<'+str(n)).\
                sort_values(by=['business_city','category_rank'],ascending=True)
    return df_temp_topn
```

```python
# To check top/ most popular category for each city
cat_rank(2)
```

```python
# Plotting top 3 most popular categories per city. For demonstration purpose limiting it to just 3 cities.

sns.set_style(style="darkgrid")
cities_list = sorted(df_temp.business_city.value_counts().index.tolist())
for city in cities_list[:3]:
    sns.barplot(x='business_categories', y='business_categories_count', data=cat_rank(4)[cat_rank(4).business_city==city])
    plt.title(str(city),fontdict= { 'fontsize': 18, 'fontweight':'bold'},pad=20)
    plt.xlabel("Business Categories",fontdict= {'fontsize': 12,'fontweight':'bold'},labelpad=20)
    plt.ylabel("Categories Counts",fontdict= {'fontsize': 12,'fontweight':'bold'},labelpad=20)
    plt.show()
```

```python
def most_reviewed_biz(df, b_city=None, top_n=10):
    if b_city:
        df=df[df['business_city']==b_city]
    
    df_business_rc = df.drop_duplicates(['business_id','business_name']).\
                            nlargest(top_n, 'business_review_count', keep='all')[['business_id',
                                                        'business_name','business_city',
                                                        'business_stars','business_review_count']]
    
    return df_business_rc
```

```python
# Best of the best

def best_biz(df, stars=5, b_city=None, topn=10):
    
    cols_list = ['business_id', 'business_name', 'business_city','business_stars', 'business_review_count']
    if b_city:
        df=df[df['business_city']==b_city]

    df_business_stars = df.drop_duplicates(cols_list).query('business_stars>='+str(stars)).\
        nlargest(topn, 'business_review_count', keep='all').\
        sort_values(by=['business_stars','business_review_count'], ascending=False)[cols_list]
    
    return df_business_stars
```
